Standard Deviation (SD) Approach:
Features with higher standard deviation tend to have more variation in their values, which can be useful for clustering as they may capture more discriminative information. You can calculate the standard deviation for each numeric feature and rank or select features based on their SD values. Features with higher SD can be prioritized for inclusion in the clustering process.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Drop rows with missing values
df.dropna(inplace=True)

# Perform K-means clustering
kmeans = KMeans(n_clusters=7, random_state=42)  # You can adjust the number of clusters as needed
kmeans.fit(X)

# Get cluster labels
cluster_labels = kmeans.labels_

# Add cluster labels to the DataFrame
df['Cluster'] = cluster_labels

# Visualize clusters (for 2D data)
plt.scatter(X['Bare_Nuclei'], X['Normal_Nucleoli'], c=cluster_labels, cmap='RdYlBu')
plt.xlabel('Bare Nuclei')
plt.ylabel('Normal Nucleoli')
plt.title('K-means Clustering')
plt.colorbar(label='Cluster')
plt.show()

# Optionally, you can inspect the centroids of the clusters
centroids = kmeans.cluster_centers_
print("Centroids:")
print(centroids)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Selecting columns for clustering
selected_cols = ['Bare_Nuclei', 'Normal_Nucleoli']
X = df[selected_cols]

# Calculate silhouette score for different values of k
silhouette_scores = []
for k in range(2, 11):  # Adjust the range of cluster numbers as needed
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Silhouette Score for K-means Clustering')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))
plt.grid(True)
plt.show()
```

```python

# Calculate silhouette score for different values of k
silhouette_scores = []
for k in range(2, 11):  # Adjust the range of cluster numbers as needed
    kmeans = KMeans(n_clusters=k, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Silhouette Score for K-means Clustering')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 11))
plt.grid(True)

# Find the optimal number of clusters (elbow point)
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 2  # Add 2 to account for starting from k=2
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'Optimal k = {optimal_k}')
plt.legend()

plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


# Selecting columns for clustering
X = df.drop(['Id', 'Class', 'Cluster'], axis=1)  # Exclude 'Id' and 'Class' columns


inertia = []

# Perform K-means clustering
kmeans = KMeans(n_clusters=2, random_state=42)  # You can adjust the number of clusters as needed
kmeans.fit(X)
inertia.append(kmeans.inertia_)

kmeans = KMeans(n_clusters=3, random_state=42)  # You can adjust the number of clusters as needed
kmeans.fit(X)
inertia.append(kmeans.inertia_)

for i in range(4,12):
    kmeans = KMeans(n_clusters=i, random_state=42)  # You can adjust the number of clusters as needed
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)


# Plot the inertia
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(inertia)+1), inertia, marker='o', linestyle='-', color='b')
plt.title('Elbow Curve for K-means Clustering (Complete Dataset)')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.xticks(range(1, len(inertia) + 1))
plt.grid(True)

plt.show()
```

```python
raw_df = df.drop(['Id', 'Class', 'Cluster'], axis = 1) 
raw_df.fillna(method ='ffill', inplace = True) 
raw_df.head(2)
```

```python
# Standardize data
scaler = StandardScaler() 
scaled_df = scaler.fit_transform(raw_df) 
  
# Normalizing the Data 
normalized_df = normalize(scaled_df) 
  
# Converting the numpy array into a pandas DataFrame 
normalized_df = pd.DataFrame(normalized_df) 
  
# Reducing the dimensions of the data 
pca = PCA(n_components = 2) 
X_principal = pca.fit_transform(normalized_df) 
X_principal = pd.DataFrame(X_principal) 
X_principal.columns = ['P1', 'P2'] 
  
X_principal
```

```python
sse=[]

kmeans = KMeans(n_clusters=2, random_state=42)  # You can adjust the number of clusters as needed
kmeans.fit(X_principal)
sse.append(kmeans.inertia_)

for i in range(3,12):
    kmeans = KMeans(n_clusters=i, random_state=42)  # You can adjust the number of clusters as needed
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

# Plot the inertia
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(inertia)+1), inertia, marker='o', linestyle='-', color='b')
plt.title('Elbow Curve for K-means Clustering (Complete Dataset)')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.xticks(range(1, len(inertia) + 1))
plt.grid(True)

plt.show()


sse
```
