```python
### In this lab assignment we will be exploring Single Linear Regression by finding Price of a Car based on its Age 
```

```python
# Step 1 : import library
import pandas as pd
```

```python
# Step 2 : import data
cars = pd.read_csv('cars.csv')
cars
```

```python
# Data preprocessing : We will be filling the null values with median values

median_age = cars['Age'].median()
cars['Age'].fillna(median_age, inplace=True)
cars
```

```python
# Step 3 : define target (y) and features (X)
```

```python
cars.columns
```

```python
y = cars['Price']
```

```python
X = cars[['Age']]
```

```python
# Step 4 : train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Step 5 : select model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Step 6 : train or fit model
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# Step 7 : predict model
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Step 8 : model accuracy
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
