```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("titanic.csv")
df
```

```python
df.drop(['Cabin','PassengerId','Name','Ticket'],axis=1,inplace=True)
```

```python
df
```

```python
print(df['Embarked'].isna().values.any())
```

```python
df
```

```python
print(df['Embarked'].isna().values.any())
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
df
```

```python
median_age = df['Age'].median()
median_embarked = df['Embarked'].median()


df['Age'] = df['Age'].fillna(median_age)
df['Embarked'] = df['Embarked'].fillna(median_age)
```

```python
X = df.drop('Survived', axis=1)
y = df['Survived']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 50)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100)
rfc = RandomForestClassifier(criterion = 'entropy', random_state = 42)

rf.fit(X_train, y_train)
rfc.fit(X_train, y_train)
```

```python
y_pred  = rf.predict(X_test)
y_pred2  = rfc.predict(X_test)
```

```python
y_pred
```

```python
y_pred2
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cm = confusion_matrix(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
cr = classification_report(y_test, y_pred)
```

```python
print(cr)
```

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

feature_importance = pd.DataFrame({'rf':rf.feature_importances_}, index = X.columns)
feature_importance.sort_values('rf',ascending = True,inplace = True)

index = np.arange(len(feature_importance))
fig, ax = plt.subplots(figsize = (18,8))
rfc_feature = ax.barh(index, feature_importance['rf'], color = 'purple', label = 'Random Forest')
ax.set(yticks = index , yticklabels = feature_importance.index)

ax.legend()
plt.show()
```
